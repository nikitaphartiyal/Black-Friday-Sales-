# THIS IS SALES PREDICTION ON BLACK FRIDAY SALES


> PROJECT OVERVIEW


This dataset contains 537,577 retail sales transactions, serving as an excellent resource for feature engineering, predictive modeling, and consumer behavior analysis. The primary objective is to predict the purchase amount (a regression task) based on customer demographics and product attributes.


---



> DATA DESCRIPTION


Total Records: 537,577

Features: 12



---


> Key Features:





User_ID – Unique identifier for each customer (5,891 distinct users).

Product_ID – Unique identifier for each product (3,623 distinct products).

Gender – Gender of the purchaser (Male/Female).

Age – Age group of the purchaser (e.g., 18-25, 26-35, etc.).

Occupation – Encoded occupation category (ranging from 0 to 20).

City_Category – Tiered city classification (A, B, or C) where the user resides.

Stay_In_Current_City_Years – Duration of the user’s residence in their current city.

Marital_Status – Binary indicator (0 = Unmarried, 1 = Married).

Product_Category_1, _2, _3 – Hierarchical product classification (numerically encoded).

Purchase (Target Variable) – The transaction amount to be predicted.
